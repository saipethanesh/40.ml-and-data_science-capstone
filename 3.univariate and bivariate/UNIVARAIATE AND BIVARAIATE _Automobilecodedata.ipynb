```python
quan=[]
qual=[]
for columnName in dataset.columns:
    #print(columnName)
    if (dataset[columnName].dtype=='object'):
       # print("qual")
        qual.append(columnName)   
    else:
        #print("quan")
        quan.append(columnName) 
```

```python
descriptive=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%",
                               "Q3:75%","99%","Q4:100%"],columns=quan)
for columnName in quan:
    descriptive[columnName]["Mean"]=dataset[columnName].mean()
    descriptive[columnName]["Median"]=dataset[columnName].median()
    descriptive[columnName]["Mode"]=dataset[columnName].mode()[0]
    descriptive[columnName]["Q1:25%"]=dataset.describe()[columnName]["25%"]
    descriptive[columnName]["Q2:50%"]=dataset.describe()[columnName]["50%"]
    descriptive[columnName]["Q3:75%"]=dataset.describe()[columnName]["75%"]
    descriptive[columnName]["99%"]=np.percentile(dataset[columnName],99)
    descriptive[columnName]["Q4:100%"]=dataset.describe()[columnName]["max"]
    
    
```

```python
descriptive=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%","99","Q4:100%","IQR",
                               "1.5Rule","Lesser","Greater","Min","Max"],columns=quan)
for coulmnName in quan:
    descriptive[coulmnName]["Mean"]=dataset[coulmnName].mean()
    descriptive[coulmnName]["Median"]=dataset[coulmnName].median()
    descriptive[coulmnName]["Mode"]=dataset[coulmnName].mode()[0]
    descriptive[coulmnName]["Q1:25%"]=dataset.describe()[coulmnName]["25%"]
    descriptive[coulmnName]["Q2:50%"]=dataset.describe()[coulmnName]["50%"]
    descriptive[coulmnName]["Q3:75%"]=dataset.describe()[coulmnName]["75%"]
    descriptive[coulmnName]["99"]=np.percentile(dataset[coulmnName],99)
    descriptive[coulmnName]["Q4:100%"]=dataset.describe()[coulmnName]["max"]
    descriptive[coulmnName]["IQR"]= descriptive[coulmnName]["Q3:75%"]-descriptive[coulmnName]["Q1:25%"]
    descriptive[coulmnName]["1.5Rule"]=1.5*descriptive[coulmnName]["IQR"]
    descriptive[coulmnName]["Lesser"]=descriptive[coulmnName]["Q1:25%"]-descriptive[coulmnName]["1.5Rule"]
    descriptive[coulmnName]["Greater"]=descriptive[coulmnName]["Q3:75%"]+ descriptive[coulmnName]["1.5Rule"]
    descriptive[coulmnName]["Min"]=dataset[coulmnName].min()
    descriptive[coulmnName]["Max"]=dataset[coulmnName].max()
```

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):
    
    vif_data = pd.DataFrame()
    vif_data["variables"] = X.columns  
    
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data
```

```python
dataset = dataset.dropna(subset=['horsepower'])
```

```python
print(calc_vif(dataset[["mpg","cylinders"]]))
```

```python
from scipy.stats import ttest_ind
usa = dataset[dataset['origin'] == 'usa']['mpg']
europe = dataset[dataset['origin'] == 'europe']['mpg']
t_statistic, p_value = ttest_ind(usa, europe)
print("Independent T-Test: t-statistic =", t_statistic, ", p-value =", p_value)
```

```python
from scipy.stats import pearsonr
mpg = dataset['mpg']
acceleration = dataset['acceleration']
correlation_coefficient, p_value = pearsonr(mpg, acceleration)
print("Pearson Correlation: correlation coefficient =", correlation_coefficient, ", p-value =", p_value)
```

```python
from scipy.stats import ttest_rel  
mpg = [21.0, 22.5, 23.0, 24.5, 25.0]
acceleration = [10.0, 9.8, 10.2, 9.7, 10.1]
t_statistic, p_value = ttest_rel(mpg, acceleration)
print("Paired T-Test (Hypothesis):", (t_statistic, p_value))
```

```python
from scipy import stats
usa_mpg = dataset[dataset['origin'] == 'usa']['mpg']
europe_mpg = dataset[dataset['origin'] == 'europe']['mpg']
japan_mpg = dataset[dataset['origin'] == 'japan']['mpg']
f_statistic, p_value = stats.f_oneway(usa_mpg, europe_mpg, japan_mpg)
print("ANOVA: F-statistic =", f_statistic, ", p-value =", p_value)
```
